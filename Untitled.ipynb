```python
# this process done beacuse this cannot be 0 in human so i have set nan 
cols = [
    'Glucose',
    'BloodPressure',
    'SkinThickness',
    'Insulin',
    'BMI'
]
df[cols] = df[cols].replace(0, np.nan)
```

```python
df.fillna(df.median(),inplace = True)
```

```python
df.drop_duplicates(inplace = True)
```

```python
plt.figure(figsize = (15,6))

for i, column in enumerate(df.columns [:-1]):
    plt.subplot(3,3,i + 1)
    sns.boxplot(y = df[column])

plt.tight_layout()
plt.show()
```

```python
x = df.drop('Outcome' , axis = 1)
y = df['Outcome']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(
    x,
    y,
    test_size = 0.2 ,
    random_state = 42,
    stratify = y
)

```

```python
# now we will gonna do feature scaling because KNN requires scaling
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 5)  #Because the developers of Scikit-learn set n_neighbors=5 as the default value when they designed KNeighborsClassifier.
knn.fit(x_train_scaled,y_train)  
```

```python
y_pred = knn.predict(x_test_scaled)
```

```python
y_pred
```

```python
from sklearn.metrics import(
            accuracy_score,
            confusion_matrix,
            classification_report,
            roc_curve,roc_auc_score
)
```

```python
accuracy = accuracy_score(y_test,y_pred)
print('Accuracy:',accuracy)
```

```python
cm = confusion_matrix(y_test,y_pred)
cm
```

```python
sns.heatmap(cm,
            fmt = 'd',
            cmap = 'Blues')
plt.xlabel('predicted')
plt.ylabel('Actual')
plt.show()
```

```python
print(classification_report(y_test,y_pred))
```

```python
accuracy = []
for k in range(1,21):
    model = KNeighborsClassifier(n_neighbors = k)
    model.fit(x_train_scaled,y_train)
    pred = model.predict(x_test_scaled)
    accuracy.append(accuracy_score(y_test,pred))
```

```python
# plot best k
plt.figure(figsize = (10,6))
plt.plot(range(1,21),
         accuracy,
         marker = 'o')
plt.xlabel('k')
plt.ylabel('Accuracy')
plt.title('Accuracy vs k')
plt.grid()
plt.show()
```

```python
best_k = accuracy.index(max(accuracy))+1
print('Best k =',best_k)
```

```python
final_model = KNeighborsClassifier(n_neighbors = best_k)
final_model.fit(x_train_scaled,y_train)
```

```python

prob = final_model.predict_proba(x_test_scaled)[:,1]
fpr,tpr,threshold = roc_curve(y_test,prob)
auc = roc_auc_score(y_test,prob)
plt.figure(figsize = (10,6))
plt.plot(fpr,tpr,label = 'AUC = %.3f'%auc)
plt.plot([0,1],[0,1],'--')
plt.xlabel('False positive Rate')
plt.ylabel('True positive Rate')
plt.legend()
plt.show()
```

```python
import joblib 
joblib.dump(final_model,'knn_diabetes.pkl')
joblib.dump(scaler,'scaler.pkl')
```

```python
# predict on new data
new_patient = [[2,120,70,25,100,30.5,0.45,35]]
new_patient = scaler.transform(new_patient)
prediction = final_model.predict(new_patient)

if prediction[0] == 1:
    print('patient is diabetic')
else:
    print("not diabetic")
```
